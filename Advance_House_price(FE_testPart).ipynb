## Missing values

### Categorical misssing values

```python
## First lets handle categorical feature which are missing

feature_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
```

```python
df[feature_nan].head()
```

```python
for feature in feature_nan:
    print('{}: {}% missing values'.format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
# Replace missing values with new level

def replace_cat_feature(df, feature_nan):
    data = df.copy()
    data[feature_nan] = df[feature_nan].fillna('Missing')
    return data

df = replace_cat_feature(df, feature_nan)
```

```python
df[feature_nan].head()
```

```python
df[feature_nan].isnull().sum()
```

## Numerical Missing Values

```python
## Now let's check the numerical feature that cointains missing values

numerical_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes !='O']
```

```python
df[numerical_nan].head()
```

```python
for feature in numerical_nan:
    print("{}: {}% missing values".format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
## Now let's fill the missing values

for feature in numerical_nan:

    median_value = df[feature].median()

    ## Create a new feature to comapre the nan values
    df[feature+'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_value, inplace=True)    
```
